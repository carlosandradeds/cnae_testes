```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from seleniumwire import webdriver as wiredriver

# Inicialize o driver do Firefox com o selenium-wire
driver = wiredriver.Firefox()

# Navegue até a página da web
driver.get('https://dados.gov.br/dados/conjuntos-dados/cadastro-nacional-da-pessoa-juridica---cnpj')  # Substitua com a URL real

# Localize o botão "Recursos" com base no atributo aria-controls
recursos_button = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, 'button[aria-controls="collapse-recursos"]'))
)

# Clique no botão "Recursos" para expandir os recursos
recursos_button.click()

# Aguarde até que os botões de download fiquem visíveis
download_buttons = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.ID, 'btnDownloadUrl'))  # Substitua com o ID real dos botões de download
)

# Você pode iterar sobre os botões de download e clicar neles um por um
for download_button in download_buttons:
    download_button.click()

    # Aguarde até que o download seja concluído
    while True:
        if "Content-Disposition" in driver.requests[-1].response.headers:
            # O download foi concluído quando o cabeçalho "Content-Disposition" estiver presente na última solicitação
            break

# Feche o navegador
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from seleniumwire import webdriver as wiredriver
import time

# Inicialize o driver do Firefox com o selenium-wire
driver = wiredriver.Firefox()

# Navegue até a página da web
driver.get('https://dados.gov.br/dados/conjuntos-dados/cadastro-nacional-da-pessoa-juridica---cnpj')  # Substitua com a URL real

# Localize o botão "Recursos" com base no atributo aria-controls
recursos_button = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, 'button[aria-controls="collapse-recursos"]'))
)

# Clique no botão "Recursos" para expandir os recursos
recursos_button.click()

# Aguarde até que os botões de download fiquem visíveis
download_buttons = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.ID, 'btnDownloadUrl'))  # Substitua com o ID real dos botões de download
)

# Você pode iterar sobre os botões de download e clicar neles um por um
for download_button in download_buttons:
    download_button.click()

# Aguarde um tempo suficiente para que todos os downloads sejam iniciados (ajuste conforme necessário)
time.sleep(10)

# Feche o navegador
driver.quit()
```

```python
import requests
from tqdm import tqdm

url = f'https://dados.rfb.gov.br/CNPJ/Empresas{n}.zip'

for n in range(10):

# Realize a solicitação GET para baixar o arquivo
response = requests.get(url, stream=True)

# Verifique se a solicitação foi bem-sucedida (código 200)
if response.status_code == 200:
    # Especifique o caminho onde você deseja salvar o arquivo
    nome_do_arquivo = 'Empresas0.zip'

    # Obtenha o tamanho total do arquivo em bytes
    total_size = int(response.headers.get('content-length', 0))

    # Crie uma barra de progresso usando o tqdm
    progress_bar = tqdm(total=total_size, unit='B', unit_scale=True)

    # Abra um arquivo local para escrever o conteúdo do arquivo
    with open(nome_do_arquivo, 'wb') as file:
        for data in response.iter_content(chunk_size=1024):
            file.write(data)
            progress_bar.update(len(data))  # Atualize a barra de progresso

    progress_bar.close()  # Feche a barra de progresso

    print(f'O arquivo foi baixado com sucesso como {nome_do_arquivo}')
else:
    print('A solicitação falhou. Status code:', response.status_code)
```
